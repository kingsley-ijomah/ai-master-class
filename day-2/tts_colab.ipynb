```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf

pipeline = KPipeline(lang_code='a')  # English by default or 'a' for all
```

```python
text = "Hello, I am Kingsley ijomah founder of codehance and AI educator"

generator = pipeline(text, voice='pf_dora')  # specify voice(s) available in Kokoro

for i, (gs, ps, audio) in enumerate(generator):
    print(f"Utterance {i+1}, graphemes: {gs}")
    display(Audio(audio, rate=24000))
    sf.write(f"{i}.wav", audio, 24000)  # Save each utterance as WAV file
```
